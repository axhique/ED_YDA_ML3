Identifying target varaible ,
Removing  rupees sign and comma from mobile_price and changing it to numeric dtype

```python
df['mobile_price']=df['mobile_price'].replace(['₹',','],'',regex=True).astype(int)
```

Bluetooth has same value 'YES' for every row , dual_sim has only one different value , hence both dropped

```python
df=df.drop(['dual_sim','bluetooth'],axis=1)
```

Removing suffixes from  numerical colums

```python
num_cols=['mob_weight','mob_height','mob_depth','mob_width','battery_power','ram','int_memory','mp_speed']

for col in num_cols:
  df[col]=df[col].str.replace('[a-zA-Z]','',regex=True)
df[num_cols]
```

Converting num_cores to numerical representation

```python
df['num_cores'].unique()
```

```python
cores={'Octa Core':8,'Quad Core':4,'Single Core':1}
df['num_cores']=df['num_cores'].map(cores)
```

Extracting OS android versions in numeric

```python
df['os'].unique()
```

```python
df['os']=df['os'].str.replace('[a-zA-Z]','',regex=True)
df['os']=[int(val.split('.')[0]) for val in df['os']]
```

```python
df['os'].unique()
```

Separating Resolution feature into two features

```python
df['resolution'].unique()
```

```python
df['resolution']=df['resolution'].replace(['\s?[pP]ixel.*','\$'],'',regex=True).replace(['\s?[Xx*×]\s?'],'x',regex=True)
df['resolution']=df['resolution'].replace(['\s?[Xx*×]\s?'],'x',regex=True)

df['resolution_1']=[val.split('x')[0] for val in df['resolution']]
df['resolution_2']=[val.split('x')[1] for val in df['resolution']]

df=df.drop('resolution',axis=1)
```

```python
df[['resolution_1','resolution_2']]
```

Separating  network column into 5g,4g,4g volte,3g,2g separate columns  for better understanding

```python
df[['network']].value_counts()
```

```python
df['network']=df['network'].replace('4G VOLTE','VOLTE',regex=True)
```

```python
net=['5G', 'VOLTE', '4G', '3G', '2G' ]
for n in net:
  df[n]= df['network'].apply(lambda x: 1 if n in x else 0)
```

```python
df[['5G', 'VOLTE', '4G', '3G', '2G' ,'network']].sample(8)
```

```python
df=df.drop('network',axis=1)
```

Extracting numeric display size from disp-size

```python
df[['disp_size']].sample(5)
```

```python
df['disp_size']=df['disp_size'].str.split('(').str[0].replace('\s?[a-zA-Z]','',regex=True)
```

```python
df[['disp_size']].sample(5)
```

Camera features

```python
df[['p_cam','f_cam']].sample(5)
```

```python
df['p_cam_max']=df['p_cam'].str.split('M').str[0]
df['f_cam_max']=df['f_cam'].str.split('M').str[0]
df['p_cam_count']=df['p_cam'].str.count('MP')
df['f_cam_count']=df['f_cam'].str.count('MP')

```

```python
df[['p_cam','f_cam','p_cam_max','p_cam_count','f_cam_max','f_cam_count']]
```

```python
df=df.drop(['p_cam','f_cam'],axis=1)
```

Mobile color

```python
df['mobile_color'].unique()
```

```python
df['mobile_color']=[color.lower().split()[-1] for color in df['mobile_color'].values]
df['mobile_color'].unique()
```

Mobile name and brand

```python
df['mobile_name'].unique()
```

```python
df['mobile_name']=df['mobile_name'].str.split(' ').str[0]
df['mobile_name'].unique()
```

```python
df.sample(10).T
```

Encoding  name and color

```python
color_map=df.groupby('mobile_color').size()/len(df)
name_map=df.groupby('mobile_name').size()/len(df)
```

```python
df['mobile_color']=df['mobile_color'].map(color_map)
df['mobile_name']=df['mobile_name'].map(name_map)
```

```python
df.sample(10).T
```

```python
df.dtypes
```

```python
intt=['p_cam_max','f_cam_max','resolution_2','resolution_1','battery_power','ram','int_memory']
flt=['mp_speed','disp_size','mob_weight','mob_height','mob_depth','mob_width']
df[intt]=df[intt].astype(int)
df[flt]=df[flt].astype(float)
```

#Model

```python
x=df.drop('mobile_price',axis=1)
y=df[['mobile_price']]
```

```python
x_train , x_test , y_train ,y_test =train_test_split(x,y,test_size=.25,random_state=42)
```

```python
x_train.shape , x_test.shape , y_train.shape ,y_test.shape
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
model.score(x_test,y_test)
```
