Removing  rupees sign and comma from mobile_price

```python
df['mobile_price']=df['mobile_price'].replace(['',','],'',regex=True)
```

Bluetooth has same value 'YES' for every row , dual_sim has only one different value , hence both dropped

```python
df=df.drop(['dual_sim','bluetooth'],axis=1)
```

Removing suffixes from  numerical colums

```python
num_cols=['mob_weight','mob_height','mob_depth','mob_width','battery_power','ram','int_memory','mp_speed']

for col in num_cols:
  df[col]=df[col].str.replace('[a-zA-Z]','',regex=True)
df[num_cols]
```

Converting num_cores

```python
df['num_cores'].unique()
```

```python
cores={'Octa Core':8,'Quad Core':4,'Single Core':1}
df['num_cores']=df['num_cores'].map(cores)
```

```python
df['os']=df['os'].str.replace('[a-zA-Z]','',regex=True)
df['os']=[int(val.split('.')[0]) for val in df['os']]
```

Dealing  resolution

```python
df['resolution']=df['resolution'].replace(['[pP]ixels','\s'],'',regex=True)
df['resolution']=df['resolution'].replace(['[Xx*]'],'x',regex=True)

df['resolution_1']=[val.split('x')[0] for val in df['resolution']]
df['resolution_2']=[val.split('x')[1] for val in df['resolution']]

df=df.drop('resolution',axis=1)
```

Separating  networks  for better understanding

```python
df[['network']].value_counts()
```

```python
df['network']=df['network'].replace('4G VOLTE','VOLTE',regex=True)
```

```python
net=['5G', 'VOLTE', '4G', '3G', '2G' ]
for n in net:
  df[n]= df['network'].apply(lambda x: 1 if n in x else 0)
df=df.drop(['network',],axis=1)
```
